### Analysis on defaulters and non defaulters based on TARGET

```python
data_A = data[data.TARGET==0]    # Dataframe with all the data related to non-defaulters
data_B = data[data.TARGET==1]    # Dataframe with all the data related to defaulters
```

## Univariate analysis on Categorical variable

```python
def univariate_application_data(var):

    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.countplot(x=var, data=data_A,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=20)
    ax1.set_xticklabels(ax1.get_xticklabels())
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(data_A))*100), (p.get_x()+0.1, p.get_height()+50))
        
    sns.countplot(x=var, data=data_B,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {var} for Defaulters',fontsize=20)    
    ax2.set_xticklabels(ax2.get_xticklabels())
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(data_B))*100), (p.get_x()+0.1, p.get_height()+50))
    
    plt.show()
```

```python
univariate_application_data('CODE_GENDER')
```

Through this plot diagram we can understand that the percentage of female applicants are higher in comparison of male applicants to the non-defaulters i.e., approx 67% of female are non defaulters whereas 33% of male are non-defaulters, whereas percentage of females are higher than males in defaulters diagram also, this means count of female applicants are more than that of male applicants.

```python
univariate_application_data('FLAG_OWN_CAR')
```

We can see that people with cars contribute 65.7% to the non-defaulters while 69.5% to the defaulters. 
While people who have car default more often, the reason could be there are simply more people without cars. Looking at the percentages in both the charts, we can conclude that the rate of default of people having car is low compared to people who don't.

## Univariate analysis on continuous variable

```python
def univariate_continuous_variable(var):

    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(15,5))
    
    sns.distplot(a=data_A[var],ax=ax1)

    ax1.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=15)
            
    sns.distplot(a=data_B[var],ax=ax2)
    ax2.set_title(f'Distribution of {var} for Defaulters',fontsize=15)    
        
    plt.show()
```

```python
univariate_continuous_variable('DAYS_EMPLOYED')
```

As we can see from the distribution graph that the peak is for the applicants who are 0 days employed are stated as non defaulters. Whereas, applicants 0 days employed for daufaulters are also having the peak.

## Bivariate Analysis of Continuous variables

```python
def bivariate(var1,var2):
    
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
    
    sns.scatterplot(x=var1, y=var2,data=data_A,ax=ax1)
    ax1.set_xlabel(var1)    
    ax1.set_ylabel(var2)
    ax1.set_title(f'{var1} vs {var2} for Non-Defaulters',fontsize=15)
    
    sns.scatterplot(x=var1, y=var2,data=data_B,ax=ax2)
    ax2.set_xlabel(var1)    
    ax2.set_ylabel(var2)
    ax2.set_title(f'{var1} vs {var2} for Defaulters',fontsize=15)
            
    plt.show()
```

```python
bivariate('AMT_CREDIT','CNT_FAM_MEMBERS')
```

We can see that the density in the lower left corner is similar in both the case, so the people are equally likely to default if the family is small and the AMT_CREDIT is low. We can observe that larger families and people with larger AMT_CREDIT default less often

```python
bivariate('AMT_GOODS_PRICE','AMT_CREDIT')
```

### Using box plot to do bivariate analysis on categorical and numeric columns


```python
def cat_con(cat, num):

        
    fig,ax = plt.subplots(1,1,figsize=(10,8))
    
    sns.boxenplot(x=cat,y = num, data=data1)
    ax.set_ylabel(f'{num}')
    ax.set_xlabel(f'{cat}')

    ax.set_title(f'{cat} Vs {num}',fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
     
    plt.show()
```

# Merging Application data and Previous Application data

```python
final_application = pd.merge(data, data1, how='inner', on=['SK_ID_CURR'])
```

#### Simple analysis on final application file

```python
final_application.shape
```

```python
final_application. info()
```

```python
def combined_data(Varx,Vary):
    
    NewData = final_application.pivot_table(values='SK_ID_CURR', index=Varx, columns=Vary, aggfunc='count')
    NewData=NewData.div(NewData.sum(axis=1),axis='rows')*100
    sns.set()
    NewData.plot(kind='bar',stacked=True,figsize=(15,5))
    plt.title(f'Effect Of {Varx} on Loan Approval')
    plt.xlabel(f'{Varx}')
    plt.ylabel(f'{Vary}%')
    plt.show()
```

```python
combined_data('FLAG_OWN_CAR', 'NAME_CONTRACT_STATUS')
```

We see that car ownership doesn't have any effect on application approval or rejection. But we saw earlier that the people who has a car has lesser chances of default. The bank can add more weightage to car ownership while approving a loan amount.
